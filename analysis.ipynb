```python
#importing necessary modules.
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import calendar
import datetime
```

```python
#upload the CSV file on which we have to work.
df=pd.read_csv("E:\Data_analysis\weatherHistory.csv.zip")
```

```python
#convert the "Formatted Date" column which is in object form to "Datetime" format
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
```

```python
#change index to 'formatted date'
df=df.set_index('Formatted Date')
```

```python
#convert the "Formaatted Column" into monthwise format from daywise format.
df=df.resample('MS').mean()
```

```python
apt=df['Apparent Temperature (C)']
```

```python
hmd=df['Humidity']
```

```python
plt.figure(figsize=(12,9))
plt.title("Analysis of Humidity and Temperature")
plt.plot(hmd,label="Average Humidity")
plt.plot(apt,label="Average Apparent Temperature")
plt.xlabel("Years")
plt.legend(loc=(1.01,0.8))
plt.show()

#This graph below shows the daywise apparent temperature from 2006 to 2016
```

```python
df1 = df[df.index.month==1]
hdt=df1['Humidity']
atm=df1['Apparent Temperature (C)']
plt.title("January Monthwise Stats",fontsize=17)
plt.plot(hdt,label="Average Humidity",marker=".")
plt.plot(atm,label="Average Apparent Temperature",marker='.')
plt.legend(loc=(1.02,0.8))
plt.show()

#As we can see that Average Apparent Temperature in January is somewhat getting affected and is most affected in 2007 and the avg temp is also decread
```
